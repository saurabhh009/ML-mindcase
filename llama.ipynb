```python
from transformers import AutoTokenizer
import transformers
import torch


model_id = "meta-llama/Llama-2-7b-chat-hf"
tokenizer = AutoTokenizer.from_pretrained(model_id)
```

```python
import PyPDF2
import transformers
import torch

# Function to extract text from PDF
# def extract_text_from_pdf(pdf_path):
#     with open(pdf_path, "rb") as file:
#         reader = PyPDF2.PdfReader(file)
#         text = ""
#         for page_num in range(len(reader.pages)):
#             text += reader.pages[page_num].extract_text()
#     return text
def extract_text_from_pdf(pdf_path):
    with open(pdf_path, "rb") as file:
        reader = PyPDF2.PdfReader(file)
        text = ""
        for page_num in range(min(2,len(reader.pages))):  # Read only the first two pages
            page = reader.pages[page_num]
            text += page.extract_text()
    return text

# Function to generate an answer based on the provided question and extracted text
def generate_answer(question, text):
    pipeline = transformers.pipeline(
        "text-generation",
        model=model_id,
        torch_dtype=torch.float16,
        device_map="auto",
        max_new_tokens=10000,
    )
    input_text = f"Question: {question}\nContext: {text}"
    answer = pipeline(input_text)[0]['generated_text']
    return answer

# Path to the PDF file
pdf_path = "/content/blade runner 2049.pdf"
# Extract text from PDF
pdf_text = extract_text_from_pdf(pdf_path)

```

```python
print(pdf_text)
```

```python
# Define a question
question = "What is the main theme of the document?"

# Generate an answer based on the question and extracted text
answer = generate_answer(question, pdf_text)

# Print the answer
print("Answer:", answer)
```
